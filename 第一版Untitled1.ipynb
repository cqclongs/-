```python
# 1. 提取所有数字（支持整数/小数）
df['area'] = df['area'].astype(str).str.extract(r'(\d+\.?\d*)')

# 2. 转换成数值类型，异常值转为 NaN
df['area'] = pd.to_numeric(df['area'], errors='coerce')

# 3. 验证结果
print(df[['area']].head())
print("数据类型：", df['area'].dtype)
```

```python
# 提取所有数字（支持整数/小数）
df['unit'] = df['unit'].astype(str).str.extract(r'(\d+\.?\d*)')

# 转为数值类型
df['unit'] = pd.to_numeric(df['unit'], errors='coerce')
```

```python
# 提取所有数字（支持整数/小数）
df['year'] = df['year'].astype(str).str.extract(r'(\d+\.?\d*)')

# 转为数值类型
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df.head()
```

```python
df=df[(df['area']<600)&(df['area']>20)]
df.head()
len(df)
```

```python
Q1=df['price'].quantile(0.25)
Q3=df['price'].quantile(0.75)
IQR=Q3-Q1
low_price=Q1-1.5*IQR
high_price=Q3+1.5*IQR
df=df[(df['price']<high_price)&(df['price']>low_price)]
```

```python
df.loc[:, 'district'] = df['address'].str.split('-').str[0]
df.head()
```

```python
df.sample(10) #看楼层格式是否都一致
```

```python
# 1. 先显式创建副本，消除 SettingWithCopyWarning
df = df.copy()
# 2. 用正则直接提取楼层类型（完全无视空格/格式）
df['floor_type'] = df['floor'].str.extract(r'^(.*?)\s*\(')
# 3. 验证结果
print(df[['floor', 'total_floor']].head())
```

```python
df.head()
```

```python
# 1. 定义直辖市列表
municipalities = ['北京', '上海', '天津', '重庆']

# 2. 新增 is_municipality 列：1=直辖市，0=非直辖市
df['is_municipality'] = df['province'].isin(municipalities).astype(int)

# 3. 查看结果（可选）
print(df[['province', 'is_municipality']].value_counts())
print(df[['city', 'province', 'is_municipality']].head())
```

```python
df.sample(5) #验证代码是否正确
```

```python
# 提取卧室数量（匹配“室”前面的数字）
df['bedrooms'] = df['rooms'].str.extract(r'(\d+)室').astype(float)

# 提取客厅数量（匹配“厅”前面的数字）
df['livingrooms'] = df['rooms'].str.extract(r'(\d+)厅').astype(float)
```

```python
# 计算楼龄（处理空值）
df['building_age'] = 2026 - df['year']
df['building_age'] = pd.to_numeric(df['building_age'], errors='coerce')
```

```python
# 定义价格区间（单位：万元）
price_bins = [0, 50, 100, 150, 200, float('inf')]
price_labels = ['0-50万', '50-100万', '100-150万', '150-200万', '200万以上']

# 生成价格分段标签
df['price_labels'] = pd.cut(
    df['price'], 
    bins=price_bins, 
    labels=price_labels, 
    right=False  # 左闭右开，避免边界值重复
)
```

```python
# 验证结果
print(df[['rooms', 'bedrooms', 'livingrooms', 'year', 'building_age', 'price', 'price_labels']].head())
```
